##Stage 2 : Caluculate cac

```python
channel_groups = df.groupby('Marketing_Channel').agg({'Marketing_Spend':'sum', 'New_Customers':'sum'})
```

```python
channel_groups
```

```python
channel_groups['CAC'] = channel_groups['Marketing_Spend'] / channel_groups['New_Customers']
```

```python
channel_groups
```

```python
channel_groups = channel_groups.sort_values(by='CAC')
```

```python
channel_groups
```

```python
df.columns
```

##Step 3: Data Visualization

```python
fig = px.bar(channel_groups,x = channel_groups.index , y = 'CAC' ,hover_name='CAC',title = 'CAC by Marketing Channel',color = channel_groups.index )
fig.show()
```

```python
fig = px.bar(channel_groups, x=channel_groups.index,
             y=['CAC', 'New_Customers'],
             title='Comparison of CAC and New Customers by Marketing Channel',
             labels={'CAC': 'Customer Acquisition Cost (CAC)', 'New_Customers': 'New Customers'},
             color_discrete_map={'CAC': 'blue', 'New_Customers': 'green'},
             barmode='group')  # barmode='group' places bars side by side

# Show the chart
fig.show()
```

```python
fig_pie = px.pie(channel_groups,names = channel_groups.index , values = 'CAC',color = channel_groups.index,hover_name = channel_groups.index,hover_data = 'CAC',title = 'CAC Distribution by Marketing Channel')
fig_pie.show()
```

##Step 4: Insights and Recommendations

###**Comparison of CAC Across Channels:**

Online Ads have the lowest CAC at approximately $99.78 per customer acquired.

Referral marketing is close behind, with a CAC of about $100.26 per customer.

Social Media has a slightly higher CAC of around $104.92 per customer.

Email Marketing has the highest CAC among the channels, at approximately $107.18 per customer.

###**Marketing Efficiency:**

Online Ads and Referral marketing seem to be the most cost-effective channels in terms of acquiring new customers.

Social Media is moderately cost-effective, while Email Marketing appears to be the least efficient in terms of cost-effectiveness.
